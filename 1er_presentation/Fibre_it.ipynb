```python
import pandas as pd
from dbfread import DBF
import os
```

```python
# √âtape 2 : Charger le fichier .dbf avec encodage correct
path = r"C:\Users\fares\OneDrive\Desktop\M√©moire\Fibre\communes_2024T4_v2.dbf"
table = DBF(path, encoding='latin-1', load=True)  # ou 'utf-8' si √ßa ne marche pas
df = pd.DataFrame(iter(table))

# √âtape 3 : V√©rifier les 5 premi√®res lignes
print("üîç Donn√©es brutes :")
print(df.head())

# √âtape 4 : Voir toutes les colonnes pour rep√©rer celle du taux de couverture fibre
print("\nüìã Colonnes disponibles :")
print(df.columns)

# √âtape 5 : R√©sum√© statistique de la variable 'couv'
print("\nüìä Statistiques de la colonne 'couv' :")
print(df["couv"].describe())

# √âtape 6 : Cr√©er la variable Fibre_it
df["Fibre_it"] = (df["couv"] >= 80).astype(int)

# √âtape 7 : V√©rifier le r√©sultat
print("\n‚úÖ Exemple apr√®s cr√©ation de Fibre_it :")
print(df[["INSEE_COM", "NOM_COM", "couv", "Fibre_it"]].head(10))

# √âtape 8 : Cr√©er une version simplifi√©e pour export
df_clean = df[["INSEE_COM", "Fibre_it"]].copy()
df_clean["annee"] = 2024

# √âtape 9 : V√©rifier le fichier final
print("\nüì¶ Fichier final pr√™t √† exporter :")
print(df_clean.head())

# √âtape 10 : Exporter
df_clean.to_csv(r"C:\Users\fares\OneDrive\Desktop\M√©moire\Fibre\fibre_it_2024.csv", index=False)
print("\n‚úÖ Export effectu√© vers fibre_it_2024.csv")
```

```python
# Dossier o√π se trouvent tous les fichiers fibre_it
folder = r"C:\Users\fares\OneDrive\Desktop\M√©moire\Fibre"

# Liste des ann√©es √† fusionner
annees = [2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Liste pour stocker les petits DataFrames
all_dfs = []

# Boucle sur chaque fichier
for annee in annees:
    file_path = os.path.join(folder, f"fibre_it_{annee}.csv")  # ‚ö° Attention √† bien v√©rifier que ce sont des .csv
    try:
        df = pd.read_csv(file_path)
        df["annee"] = annee  # Juste pour √™tre s√ªr d'avoir l'ann√©e
        all_dfs.append(df)
        print(f"‚úÖ Fichier charg√© : {file_path}")
    except Exception as e:
        print(f"‚ö†Ô∏è Erreur avec {file_path} : {e}")

# Fusionner tous les fichiers ensemble
panel_fibre = pd.concat(all_dfs, ignore_index=True)
panel_fibre = panel_fibre.sort_values(by=["INSEE_COM", "annee"]).reset_index(drop=True)

# V√©rifier rapidement
print(panel_fibre.head())

# Sauvegarder le panel final
output_path = os.path.join(folder, "panel_fibre_it_2018_2024.csv")
panel_fibre.to_csv(output_path, index=False)

print(f"\n‚úÖ Panel final export√© vers : {output_path}")
```
